## **01 If --> Check Ganjil atau Genap**

```python
bilangan = 2.8


if(bilangan % 2) ==1: #suatu kondisi jika
    print("Bilangan Ganjil")
    bilangan = bilangan**2
    print(bilangan)
elif (bilangan % 2) ==0: #kondisi
    print("Bilangan Genap")

else: #kondisi apabila tidak terpenuhi
    print("Tidak termasuk bilangan")

```

### **Program Menghitung Usia Rata-rata Keluarga**

```python
jumlah_anggota = int(input("Masukkan jumlah anggota keluarga: ")) #jumlah anggota keluarga dari user

total_usia = 0 #Inisialisasi total usia

for i in range(1, jumlah_anggota + 1): #Input usia tiap anggota
    usia = int(input(f"Masukkan usia anggota keluarga ke-{i}: "))
    total_usia += usia

rata_rata = total_usia / jumlah_anggota #Hitung rata-rata

# Tampilkan hasil
print(f"\nRata-rata usia keluarga adalah: {rata_rata:.2f} tahun")
```
