Konversi Suhu

```python
def celsius_ke_fahrenheit(c):
    "Mengubah suhu dari Celcius ke Fahrenheit"
    return (c * 9/5) + 32

def celsius_ke_kelvin(c):
    "Mengubah suhu dari Celcius ke Kelvin"
    return c + 273.15
c = float(input("Masukkan suhu dalam Celcius: "))
f = celsius_ke_fahrenheit(c)
k = celsius_ke_kelvin(c)

print(f"Suhu dalam Celcius: {c}")
print(f"Suhu dalam Fahrenheit: {f}")
print(f"Suhu dalam Kelvin: {k}")
```

Diskon Belanja

```python
def hitung_total(harga, jumlah, diskon):
    total_awal = harga * jumlah

    potongan = (diskon / 100) * total_awal

    total_akhir = total_awal - potongan

    return total_akhir

harga = int(input("Masukkan harga barang: "))
jumlah = int(input("Masukkan jumlah barang: "))
diskon = float(input("Masukkan diskon (%): "))

total = hitung_total(harga, jumlah, diskon)
print("Total belanja: ", int(total))
```
